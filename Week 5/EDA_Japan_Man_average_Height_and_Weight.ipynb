```python
#  membaca dataset CSV
path_to_file = '/content/drive/My Drive/Dataset/man.csv'
man = pd.read_csv(path_to_file)
man
```

```python
# Mengecek missing values
missing_values = man.isnull().sum()

# Menampilkan jumlah missing values untuk setiap kolom
print(missing_values)
```

```python
# Melihat ukuran data
ukuran_data = man.shape
print("Ukuran data (baris, kolom):", ukuran_data)
```

```python
sns.boxplot(x=man['year'])
```

```python
sns.boxplot(x=man['height_average'])
```

```python
sns.boxplot(x=man['height_standard deviation'])
```

```python
sns.boxplot(x=man['body weight _average'])
```

```python
sns.boxplot(x=man['body weight _standard deviation'])
```

```python
man
```

```python
man.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# Pilih fitur numerik yang ingin dianalisis
numerical_features = ['height_average', 'height_standard deviation', 'body weight _average', 'body weight _standard deviation']

# Hitung matriks korelasi antar fitur numerik
correlation_matrix = man[numerical_features].corr()

# Buat heatmap korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap Korelasi Antar Fitur Numerik')
plt.show()
```

```python
#mengamati hubungan antar fitur numerik dengan fungsi pairplot()
sns.pairplot(man, diag_kind = 'kde')
```

```python
# Pilih fitur numerik yang ingin dianalisis
numerical_features = ['height_average', 'height_standard deviation', 'body weight _average', 'body weight _standard deviation']
categorical_feature = 'category'

# Visualisasi pair plot dengan pewarnaan berdasarkan 'category'
sns.pairplot(man, hue=categorical_feature)
plt.show()
```

Dari tahapan-tahapan yang dilakukan ketika melakukan exploratory data analys kita dapat mengetahui apa saja feature yang ada. Apakah feature tersebut termasuk categorical feature atau numerical feature.

Ketika melakukan exploratory data anlalys kita terkadang menmukan dataset memiliki missing value pada suatu feture, dan ada banyak cara untuk mengatasinya seperti menghilangkan feature tersebut jika memang tidak terlalu berpengaruh hingga memasukkna nilai rata-rata atau mean untuk mengisi missing value tersebut.

Kita juga dapat melihat distribusi atau korelasi setiap feature dengan melakukan visualisasi data sperti melihat distribusi feature pada histogram atau melihat korelasi dengan heatmap atau trend pada pairplot
